```python
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt

# Configure matplotlib for better inline plotting in notebooks
%matplotlib inline
```

```python
# --- EDIT THIS LINE ---
# Replace with the path to your .npz file
npz_file_path = Path("/home/aleph/Desktop/emb_wav2vec2/REDLAT_BE00128_Phonological.npz")

# Check if the file exists to avoid errors
if not npz_file_path.is_file():
    print(f"File not found! Please check the path: {npz_file_path}")
else:
    print(f"File found. Ready to load: {npz_file_path.name}")
```

```python
# Load the .npz file. It acts like a dictionary.
data = np.load(npz_file_path)

print(f"--- Inspecting: {npz_file_path.name} ---")
print(f"Arrays found in the file: {data.files}\n")

# Iterate through each array in the file
for key in data.files:
    array = data[key]
    print(f"Array '{key}':")
    print(f"  - Shape: {array.shape}  (Time Steps, Embedding Dimension)")
    print(f"  - Data Type: {array.dtype}")

print("\n--- Inspection Complete ---")
```

```python
# Choose a layer to visualize (e.g., the last one)
layer_to_visualize = 'layer_0'

if layer_to_visualize in data.files:
    embedding_matrix = data[layer_to_visualize]

    # The matrix shape is (time_steps, embedding_dim)
    # We transpose it for a more standard visualization (embedding_dim, time_steps)
    plt.figure(figsize=(15, 5))
    plt.imshow(embedding_matrix.T, aspect='auto', origin='lower', cmap='viridis')
    
    plt.colorbar(label='Activation Value')
    plt.title(f'Visualization of "{layer_to_visualize}"')
    plt.xlabel('Time Steps')
    plt.ylabel('Embedding Dimension')
    plt.show()
else:
    print(f"Layer '{layer_to_visualize}' not found in the file.")
```
